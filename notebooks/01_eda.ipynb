### Visuals

```python
sns.countplot(data = X, x='buying')
plt.title('Distribution of Buying Prices')
plt.show()
```

```python
sns.countplot(data = X, x= 'safety')
plt.title('Distribution of safety')
plt.show()
```

```python
sns.countplot(data= X, x = 'doors')
plt.title('Distribution of doors')
plt.show()  
```

```python
sns.countplot(data=X, x='maint')
```

##### Grouped Bar Chart

```python
sns.countplot(x='buying', hue='maint', data=X)
plt.title("Buying vs Maintenance")
plt.show()
##### Grouped Bar Chart
```

##### Mosaic Plot

```python
from statsmodels.graphics.mosaicplot import mosaic

mosaic(X, ['buying', 'doors'])
plt.title("Mosaic Plot: Buying vs Doors")
plt.show()
```
